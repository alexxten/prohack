# Обработка NaN

# 1. Удаляем >70% Nan, заполняем средним

```
data = copy.deepcopy(initial_data)
```

## 1.1. Удаление столбцов 

```
def process_NaN_columns(
    *, 
    data: pd.DataFrame, 
    column_name: str, 
    percentage_value: int = 70,
) -> pd.DataFrame:
    df = copy.deepcopy(data)
    observations, _ = df.shape
    nan_values = df[column_name].isna().sum()
    print(f'{column_name} has {nan_values} Nan')
    if nan_values/observations*100 > percentage_value:
      df = df.drop(column_name, axis=1)
      print(f'----{column_name} with {nan_values} NaN was dropped----')
    return df
```

```
new_data = data
for col in data.columns:
  new_data = process_NaN_columns(data=new_data, column_name=col)
```

## 1.2. Удаление строк

```
data_1_2 = copy.deepcopy(new_data)
```

```
data_1_2 = data_1_2.dropna(axis = 1, thresh=len(data_1_2.columns)/2)
data_1_2.shape
```

## 1.3. Заполнение Nan

```
data_1_3 = copy.deepcopy(data_1_2)
```

```
data_1_3.describe()
```

```
data_1_3 = data_1_3.fillna(data_1_3.mean())
data_1_3.to_csv('train_del_70_fill_mean.csv', index=False)
```

# 2. Удаляем >70% Nan, заполняем IterativeImputer

```
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```
data_2 = copy.deepcopy(data_1_2)
data_2 = data_2.drop('galaxy', axis=1)
```

```
imputer = IterativeImputer(random_state=0)
imputer.get_params()
data_2_1 = imputer.fit_transform(data_2)
```

```
data_2_1 = pd.DataFrame(data_2_1)
data_2_1.columns = data_2.columns
data_2_1.head()
```

```
data_2_1.to_csv('train_del_70_iterative_imp.csv', index=False)
```
